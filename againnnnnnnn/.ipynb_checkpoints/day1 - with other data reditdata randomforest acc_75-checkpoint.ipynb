```python
fina=[]
def conver(data):
    for val in data:
        lm = WordNetLemmatizer()
        stopword=set(stopwords.words("english"))
        wor=re.sub("[^A-Za-z]",' ',str(val))
        wor=wor.lower() 
        wor=wor.split()
        wor = [lm.lemmatize(word) for word in wor if word not in set(stopwords.words('english'))]    
        tex=[]
        for ijx in wor:
            if ijx not in stopword:
                tex.append(lm.lemmatize(ijx))
        fina.append(" ".join(tex))
    print(fina)
    print(len(fina))    
    
train_fina=[]
def train_conver(data):
    for val in data:
        lm = WordNetLemmatizer()
        stopword=set(stopwords.words("english"))
        wor=re.sub("[^A-Za-z]",' ',str(val))
        wor=wor.lower() 
        wor=wor.split()
        wor = [lm.lemmatize(word) for word in wor if word not in set(stopwords.words('english'))]    
        tex=[]
        for ijx in wor:
            if ijx not in stopword:
                tex.append(lm.lemmatize(ijx))
        train_fina.append(" ".join(tex))
    print(train_fina)
    print(len(train_fina))
    
    
```

```python
conver(xtrain)
# print(len(xtrain))
# for i in xtrain:
#     print(i)
train_conver(xtest)
```
