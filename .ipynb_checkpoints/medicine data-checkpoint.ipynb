```python
prepro=[]
def prep(data):
    for val in data:
        lm = WordNetLemmatizer()
        stopword=set(stopwords.words("english"))
        wor=re.sub(r'http\S+', '', val, flags=re.MULTILINE)
        wor=re.sub("[^A-Za-z]",' ',str(wor))
        wor=wor.lower() 
        wor=wor.split()
        wor = [lm.lemmatize(word) for word in wor if word not in set(stopwords.words('english'))]    
        tex=[]
        for ijx in wor:
            if ijx not in stopword:
                tex.append(lm.lemmatize(ijx))
        prepro.append(" ".join(tex))
    print(prepro)
    print(len(prepro))    
```
